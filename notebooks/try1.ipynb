```python
df.groupby(['year', 'vetstat', 'empstat'])['perwt'].sum().to_frame()
```

```python
vet_group = df[df['empstat'] != 'Not in labor force'].groupby(['year', 'vetstat', 'empstat']).agg({'perwt': 'sum'})
# Change: groupby state_office and divide by sum
vet_pcts = vet_group.groupby(level=[0,1]).apply(lambda x: x / float(x.sum()))
```

```python
vet_pcts.style.format('{:.2%}')
```

```python
vet_gg = df.groupby(['year', 'vetstat'])['perwt'].sum()
vet_gg_pctg = vet_gg.groupby(level=0).apply(lambda x: x / x.sum())
```

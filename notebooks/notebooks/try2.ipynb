```python
import pandas as pd
```

```python
df = pd.read_stata("../../data/raw/pandas_wkshp.dta")

df.head()
```

```python
counties = df.groupby(['year', 'countyfips', 'empstat'])['perwt'].sum()

counties_pctgs = counties.groupby(['year', 'countyfips']).apply(lambda x: x/x.sum())
```

```python
counties_pctgs.to_frame().style.format("{:.2%}")
```
